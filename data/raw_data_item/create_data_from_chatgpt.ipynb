```python
def ask_chatgpt(instruction):
    messages = [
        {
            'role': 'system',
            'content': 'You are a helpful assistant.'
        },
        {
            'role': 'user',
            'content': instruction
        }
    ]
    response = openai.ChatCompletion.create(
        model='gpt-3.5-turbo',
        messages=messages,
    )
    return response.choices[0].message.content
```

```python
message = "请给出下面模板的五个不同表述，使得语义不变，但表述形式上具有多样性，答案用json格式给出（注意：只需要给出模板，不需要对<full name>等内容填空）：{\"{}\"}"
```

```python
result = ask_chatgpt(message)
```
